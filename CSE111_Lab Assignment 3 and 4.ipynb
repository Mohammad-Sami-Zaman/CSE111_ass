```python
#TASK no 1
class Patient:
    def __init__(self, name, age, weight, height):
        self.n = name
        self.a = age
        self.w = weight
        self.h = height
        self.hm = self.h / 100
        self.bmi = self.w / (self.hm ** 2)
    def printDetails(self):
        print('Name: ', self.n)
        print('Age: ', self.a)
        print('weight: ', self.w, 'kg')
        print('height: ', self.h, 'cm')
        print('BMI: ', self.bmi)
p1 = Patient("A", 55, 63.0, 158.0)
p1.printDetails()
print("====================")
p2 = Patient("B", 53, 61.0, 149.0)
p2.printDetails()
```

```python
#TASK no 2
class Shape:
    def __init__(self, shape, p1, p2):
        self.shape = shape
        self.p1 = p1
        self.p2 = p2
    def area(self):
        if self.shape == 'Triangle':
            self.a = 0.5 * self.p1 * self.p2
            print('Area: ', self.a)
        elif self.shape == 'Square':
            self.a = self.p1 * self.p2
            print('Area: ', self.a)
        elif self.shape == 'Rhombus':
            self.a = 0.5 * self.p1 * self.p2
            print('Area: ', self.a)
        elif self.shape == 'Rectangle':
            self.a = self.p1 * self.p2
            print('Area: ', self.a)
        else:
            print('Area: Shape unknown')
triangle = Shape("Triangle",10,25)
triangle.area()
print("==========================")
square = Shape("Square",10,10)
square.area()
print("==========================")
rhombus = Shape("Rhombus",18,25)
rhombus.area()
print("==========================")
rectangle = Shape("Rectangle",15,30)
rectangle.area()
print("==========================")
trapezium = Shape("Trapezium",15,30)
trapezium.area()
```

```python
#TASK no 3
class Calculator:
    def __init__(self):
        print('Calculator is ready!')
    def calculate(self, v1, v2, oparate):
        self.v1 = v1 
        self.v2 = v2
        self.o = oparate
        if self.o == '+':
            val = self.v1 + self.v2
        elif self.o == '-':
            val = self.v1 - self.v2
        elif self.o == '*':
            val = self.v1 * self.v2
        elif self.o == '/':
            val = self.v1 / self.v2
        self.val = val
        return val
    def showCalculation(self):
        print(f'{self.v1} {self.o} {self.v2} = {self.val}')
c1 = Calculator()
print("==================")
val = c1.calculate(10, 20, '+')
print("Returned value:", val)
c1.showCalculation()
print("==================")
val = c1.calculate(val, 10, '-')
print("Returned value:", val)
c1.showCalculation()
print("==================")
val = c1.calculate(val, 5, '*')
print("Returned value:", val)
c1.showCalculation()
print("==================")
val = c1.calculate(val, 16, '/')
print("Returned value:", val)
c1.showCalculation()
```

```python
#TASK no 4
class Programmer:
    def __init__(self, name, language, experience):
        self.n = name
        self.l = language
        self.e = experience
        self.a = None
    def addExp(self, add):
        self.a = add
        self.e = self.e + self.a
        return self.e
    def printDetails(self):
      if self.a == None:
        print('Horray! A new programmer is born')
      else:
        print('Updating experience of', self.n)
      print('Name: ', self.n)
      print('Language: ', self.l)
      print('Experience: ', self.e, 'years')
p1 = Programmer("Ethen Hunt", "Java", 10)
p1.printDetails()
print('--------------------------')
p2 = Programmer("James Bond", "C++", 7)
p2.printDetails()
print('--------------------------')
p3 = Programmer("Jon Snow", "Python", 4)
p3.printDetails()
p3.addExp(5)
p3.printDetails()
```

```python
#Task no 6
class Customer:
    def __init__(self, name):
        self.n = name
    def greet(self, name_1 = None):
        if name_1 == None:
            print('Hello!')
        else:
            print(f"Hello {name_1}!")
    def purchase(self, *args):
        print(f"{self.n}, you purchased {len(args)} item(s)")
        for elem in args:
            print(elem)
customer_1 = Customer("Sam")
customer_1.greet()
customer_1.purchase("chips", "chocolate", "orange juice")
print("-----------------------------")
customer_2 = Customer("David")
customer_2.greet("David")
customer_2.purchase("orange juice")
```

```python
#TASK no 7
class Cat:
    def __init__(self, name = None, action = None):
        if name == None and action == None:
            self.n = 'White'
            self.a = 'sitting'
        elif name != None and action == None:
            self.n = name
            self.a = 'sitting'
        else:
            self.n = name
            self.a = action
    def changeColor(self, color = None):
        if color != None:
            self.n = color
    def printCat(self):
        print(f'{self.n} cat is {self.a}')
c1 = Cat()
c2 = Cat("Black")
c3 = Cat("Brown", "jumping")
c4 = Cat("Red", "purring")
c1.printCat()
c2.printCat()
c3.printCat()
c4.printCat()
c1.changeColor("Blue")
c3.changeColor("Purple")
c1.printCat()
c3.printCat()
```

```python
#TASK NO 9
class Batsman:
  def  __init__(self, *args):
    if len(args) == 2:
        self.n  = 'New Batsman'
        self.run = args[0]
        self.ball = args[1]
    elif len(args) == 3:
        self.n  = args[0]
        self.run = args[1]
        self.ball = args[2]
  def setName(self, n1):
      self.n = n1
  def printCareerStatistics(self):
      print('Name:', self.n)
      print(f"Runs Scored {self.run}, Balls Faced: {self.ball}")
  def battingStrikeRate(self):
      self.s = self.run / self.ball * 100
      print(self.s)
b1 = Batsman(6101, 7380)
b1.printCareerStatistics()
print("============================")
b2 = Batsman("Liton Das", 678, 773)
b2.printCareerStatistics()
print("----------------------------")
print(b2.battingStrikeRate())
print("============================")
b1.setName("Shakib Al Hasan")
b1.printCareerStatistics()
print("----------------------------")
print(b1.battingStrikeRate())
```

```python
# TASK no 10
class Author:
    def __init__(self, *args):
        self.book = []
        if len(args) == 0:
            self.name = "Default"
        elif len(args) == 1:
            self.name = args[0]
        else:
            self.name = args[0]
            for elem in range(1, len(args), 1):
                self.book.append(args[elem])

    def addBooks(self, *args):
        for e in args:
            self.book.append(e)

    def changeName(self, name):
        self.name = name

    def printDetails(self):
        print('Author Name: ', self.name)
        print('-----------')
        print('List of Books:')
        for e in self.book:
            print(e)

auth1 = Author('Humayun Ahmed')
auth1.addBooks('Deyal', 'Megher Opor Bari')
auth1.printDetails()
print("===================")
auth2 = Author()
print(auth2.name)
auth2.changeName('Mario Puzo')
auth2.addBooks('The Godfather', 'Omerta', 'The Sicilian')
print("===================")
auth2.printDetails()
print("===================")
auth3 = Author('Paolo Coelho', 'The Alchemist', 'The Fifth Mountain')
auth3.printDetails() 
```

```python
#TASK no 11
class TaxiLagbe:
    def __init__(self, num, place):
        self.num = num
        self.place = place
        self.passenger = []
        self.total = 0

    def addPassenger(self, *args):
        if len(self.passenger) < 4:  
            for elem in args:
                var = elem.split('_')
                self.passenger.append(var[0])

                print(f'Dear {var[0]}! Welcome to TaxiLagbe')
                self.total = self.total + int(var[1])
        else:
            print('Taxi Full! No more passengers can be added.')

    def printDetails(self):
        print('Trip info for Taxi number: ', self.num)
        print('This taxi can cover only ', self.place, "area.")
        print('Total passenger: ', len(self.passenger))
        print("Passenger lists:")
        for i in range(len(self.passenger)):
            if i < len(self.passenger) - 1:
                print(self.passenger[i], end= ',')
            else:
                print(self.passenger[i], "")
        print('Total collected fare: ', self.total, 'Taka')
taxi1 = TaxiLagbe('1010-01', 'Dhaka')
print('-------------------------------')
taxi1.addPassenger('Walker_100', 'Wood_200')
taxi1.addPassenger('Matt_100')
taxi1.addPassenger('Wilson_105')
print('-------------------------------')
taxi1.printDetails()
print('-------------------------------')
taxi1.addPassenger('Karen_200')
print('-------------------------------')
taxi1.printDetails()
print('-------------------------------')
taxi2 = TaxiLagbe('1010-02', 'Khulna')
taxi2.addPassenger('Ronald_115')
taxi2.addPassenger('Parker_215')
print('-------------------------------')
taxi2.printDetails()
```

```python
#TASK no 12
class Account:
    def __init__(self, name = "Defalt Account", balance = 0.0):
        self.name = name
        self.balance = balance

    def details(self):
        print(self.name)
        print(self.balance, end = '')
        return ''
    def withdraw(self, amount):
        if self.balance <= amount:
            print('Sorry, Withdraw unsuccessful! The account balance after deducting withdraw amount is equal to or less than minimum.')
        else:
            self.balance = self.balance - amount
            print('Withdraw successful! New balance is: ', self.balance)
a1 = Account()
print(a1.details())
print("------------------------")
a1.name = "Oliver"
a1.balance = 10000.0
print(a1.details())
print("------------------------")
a2 = Account("Liam")
print(a2.details())
print("------------------------")
a3 = Account("Noah",400)
print(a3.details())
print("------------------------")
a1.withdraw(6930)
print("------------------------")
a2.withdraw(600)
print("------------------------")
a1.withdraw(6929)
```

```python
#TASK no 13
class StudentDatabase:
    def __init__(self, name, id):
        self.name = name
        self.id = id

    def calculateGPA(self, course, semester):
        self.course = course
        self.semester = semester
        self.grades = {self.semester : self.course}
        self.a = []
        self.sub = []
        self.rel = []
        
        total = 0
        for elem in self.course:
            a = elem.split(': ')
            self.a.append(a)
        for itaretion in self.a:
            self.sub.append(itaretion[0])
            self.rel.append(itaretion[1])
        for i in self.rel:
            total = total + float(i)
        self.gpa = total / len(self.rel)
         
    def printDetails(self):
        print('Name: ', self.name)
        print('ID: ', self.id)
        print(f'Course taken in {self.semester}:') 
        for e in self.sub:
            print(e) 
        print('GPA: ', self.gpa)  
s1 = StudentDatabase('Pietro', '10101222')
s1.calculateGPA(['CSE230: 4.0', 'CSE220: 4.0', 'MAT110: 4.0'], 'Summer2020')
s1.calculateGPA(['CSE250: 3.7', 'CSE330: 4.0'], 'Summer2021')
print(f'Grades for {s1.name}\n{s1.grades}')
print('------------------------------------------------------')
s1.printDetails()
s2 = StudentDatabase('Wanda', '10103332')
s2.calculateGPA(['CSE111: 3.7', 'CSE260: 3.7', 'ENG101: 4.0'], 'Summer2022')
print('------------------------------------------------------')
print(f'Grades for {s2.name}\n{s2.grades}')
print('------------------------------------------------------')
s2.printDetails()
```
