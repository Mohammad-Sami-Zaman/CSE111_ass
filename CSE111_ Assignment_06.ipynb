```python
#TASK no 2
class Assassin:
    total = 0
    def __init__(self, name, success):
        self.n = name
        self.s = success
        Assassin.total = Assassin.total + 1
        self.t = Assassin.total
    def printDetails(self):
        print("Name: ", self.n)
        print(f"Success: {self.s}%")
        print("Total number of Assassin: ", self.t)
    @classmethod
    def failureRate(cls, name, fail_rate):
        cls.n = name
        cls.s = 100 - fail_rate
        return Assassin( cls.n , cls.s )
    @classmethod
    def failurePercentage(cls, name, fail_percent):
        cls.n = name
        var = fail_percent[0:len(fail_percent)-1]
        cls.s = 100 - int(var)
        return Assassin(cls.n, cls.s)
john_wick = Assassin('John Wick', 100)
john_wick.printDetails()
print('================================')
nagisa = Assassin.failureRate("Nagisa", 20)
nagisa.printDetails()
print('================================')
akabane = Assassin.failurePercentage("Akabane", "10%")
akabane.printDetails()
```

```python
# TASK no 3
class Passenger:
    count = 0
    fare = 0
    def __init__(self, name):
        self.n = name
        self.fare = 450
        Passenger.count = Passenger.count + 1
    def set_bag_weight(self, weight):
        self.w = weight
        if self.w <= 20:
            Passenger.fare = Passenger.fare + self.fare
        elif 20 < self.w <= 50 :
            self.fare = self.fare + 50
            Passenger.fare = Passenger.fare + self.fare
        elif self.w > 50:
            self.fare = self.fare + 100
            Passenger.fare = Passenger.fare + self.fare
        return Passenger.fare ,Passenger.count
    def printDetail(self):
        print('Name: ', self.n)
        print('Bus fare: ', self.fare)

print('Total Passenger:', Passenger.count)
p1 = Passenger('Jack')
p1.set_bag_weight(90)
p2 = Passenger('Carol')
p2.set_bag_weight(10)
p3 = Passenger('Mike')
p3.set_bag_weight(25)
print("=========================")
p1.printDetail()
print("=========================")
p2.printDetail()
print("=========================")
p3.printDetail()
print("=========================")
print('Total Passenger:', Passenger.count)       
```

```python
#Task no 5
class Employee:
    def __init__(self, name, ex):
        self.name = name
        self.workingPeriod = ex
    @classmethod
    def employeeByJoiningYear(cls, name, year):
        cls.n = name
        cls.ex = 2022 - year
        return Employee(cls.n, cls.ex)
    @staticmethod
    def experienceCheck(ex , chr):
        if ex < 3:
            if chr.lower() == 'male':
                return "He is not experienced"
            else:
                return 'She is not experienced'
        else:
            if chr.lower() == 'male':
              return "He is experienced"
            else:
                return "She is experienced"
employee1 = Employee('Dororo', 3)
employee2 = Employee.employeeByJoiningYear('Harry', 2016)
print(employee1.workingPeriod)
print(employee2.workingPeriod)
print(employee1.name)
print(employee2.name)
print(Employee.experienceCheck(2, "male"))
print(Employee.experienceCheck(3, "female"))
```

```python
# TASK no 7
class Cat:
    Number_of_cats = 0
    def __init__(self, c = None, p = None):
       if c == None and p == None:  
         self.c = "White"
         self.p = 'sitting'
       else:
           self.c = c
           self.p = p
       Cat.Number_of_cats = Cat.Number_of_cats + 1
    @classmethod
    def no_parameter(cls):
        cls.c = "White"
        cls.p = 'sitting'
        return Cat(cls.c , cls.p)
    @classmethod
    def first_parameter(cls, c):
        cls.c = c
        cls.p = 'sitting'
        return Cat(cls.c , cls.p)
    @classmethod
    def second_parameter(cls, p):
        cls.c = 'Grey'
        cls.p = p
        return Cat(cls.c , cls.p)
    def changeColor(self, c):
        self.c = c
    def printCat(self):
        print(f'{self.c} cat is {self.p} ')
print("Total number of cats:", Cat.Number_of_cats)
c1 = Cat.no_parameter()
c2 = Cat.first_parameter("Black")
c3 = Cat("Brown", "jumping")
c4 = Cat("Red", "purring")
c5 = Cat.second_parameter("playing")
print("=======================")
c1.printCat()
c2.printCat()
c3.printCat()
c4.printCat()
c5.printCat()
c1.changeColor("Blue")
c3.changeColor("Purple")
c1.printCat()
c3.printCat()
print("=======================")
print("Total number of cats:", Cat.Number_of_cats)        
```

```python
#TASK no 8
class Cylinder:
  radius, height = 5, 18

  def __init__(self, rad, ht):
    self.radius = rad
    self.height = ht
    print(f'''Default radius= {Cylinder.radius} and height= {Cylinder.height}.
Updated: radius={self.radius} and height={self.height}''')
    #saved
    Cylinder.radius, Cylinder.height = rad, ht
    
  @classmethod
  def swap(cls, height, radius):
    return Cylinder(radius, height)

  @classmethod
  def changeFormat(cls, stri):
    x = stri.split('-')
    return Cylinder(float(x[0]), float(x[1]))
  
  @staticmethod
  def area(rad, ht):
    area = (2*3.1416*(rad**2)) + (2*3.1416*rad*ht)
    print(f"Area: {area}")
  
  @staticmethod
  def volume(rad, ht):
    volume = (3.1416 * (rad**2) * ht)
    print(f"Volume:{volume}")
c1 = Cylinder(0,0)
Cylinder.area(c1.radius,c1.height)
Cylinder.volume(c1.radius,c1.height)
print("===============================")
c2 = Cylinder.swap(8,3)
c2.area(c2.radius,c2.height)
c2.volume(c2.radius,c2.height)
print("===============================")
c3 = Cylinder.changeFormat("7-13")
c3.area(c3.radius,c3.height)
c3.volume(c3.radius,c3.height)
print("===============================")
Cylinder(0.3,5.56).area(Cylinder.radius,Cylinder.height)
print("===============================")
Cylinder(3,5).volume(Cylinder.radius,Cylinder.height)
```
