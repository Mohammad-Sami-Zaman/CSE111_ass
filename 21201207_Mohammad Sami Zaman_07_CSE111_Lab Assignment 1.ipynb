**Sample Input** 

‘madam’ 

‘hello’ 

‘nurses run

**Sample Output** 

Palindrome 

Not a palindrome

```python
def palindrom_check(word):
    word.strip()
    edited_word = ''
    for itaretion in word:
        if itaretion != ' ':
            edited_word = edited_word + itaretion

    if edited_word[::-1] == edited_word:
        print("Palindrome")
    else:
        print("Not Palindrome")
    return('')
word = input('enter Word: ')
print(palindrom_check(word))
```
