```python
# Task no 01
class Marks:
    def __init__(self, mark):
        self.mark = mark
    def __add__(self, other):
        temp = self.mark + other.mark
        return Marks(temp)

Q1 = Marks(int(input("Quiz 1 (out of 10): ")))
Q2 = Marks(int(input("Quiz 2 (out of 10): ")))
Lab = Marks(int(input("Lab (out of 30): ")))
Mid = Marks(int(input("Mid (out of 20): ")))
Final = Marks(int(input("Final (out of 30): ")))
total = Q1 + Q2 + Lab + Mid + Final
print("Total marks: {}".format(total.mark))
```

```python
class Marks:
    def __init__(self, mark):
        self.mark = mark
    def __add__(self, other):
        temp = self.mark + other.mark
        return Marks(temp)
    def __sub__(self, other):
        temp = self.mark - other.mark
        return Marks(temp)

Q1 = Marks(int(input("Quiz 1 (out of 10): ")))
Q2 = Marks(int(input("Quiz 2 (out of 10): ")))
Lab = Marks(int(input("Lab (out of 30): ")))
Mid = Marks(int(input("Mid (out of 20): ")))
Final = Marks(int(input("Final (out of 30): ")))
total = Q1 + Q2 + Lab + Mid - Final
print("Total marks: {}".format(total.mark))
```

```python
#TASK no 4
class Color:
    def __init__(self, clr):
        self.clr = clr
    def __add__(self, other):
        if (self.clr == 'blue' and other.clr == 'red') or (self.clr == 'red' and other.clr == 'blue'):
            temp = 'Violet'
        elif (self.clr == 'blue' and other.clr == 'yellow') or (self.clr == 'yellow' and other.clr == 'blue'):
            temp = 'Green'
        elif (self.clr == 'yellow' and other.clr == 'red') or (self.clr == 'red' and other.clr == 'yellow'):
            temp = 'Orange'
        return Color(temp)
C1 = Color(input("First Color: ").lower())
C2 = Color(input("Second Color: ").lower())
C3 = C1 + C2
print("Color formed:", C3.clr)
```

```python
# TASK no 06
class Triangle:
    def __init__(self, base, height):
        self.__b = base
        self.__h = height
    def setBase(self, b):
        self.__b = b
    def getBase(self):
        return(self.__b)
    def setHeight(self, h):
        self.__h = h  
    def getHeight(self):
        return(self.__h)
    def area(self):
        temp = 0.5 * self.__b * self.__h
        return temp
    def __sub__(self, other):
        val = self.__b - other.__b
        var = self.__h - other.__h
        return Triangle(val, var)
t1 = Triangle(10, 5)
print("First Triangle Base:" , t1.getBase())
print("First Triangle Height:" , t1.getHeight())
print("First Triangle area:" ,t1.area())
 
t2 = Triangle(5, 3)
print("Second Triangle Base:" , t2.getBase())
print("Second Triangle Height:" , t2.getHeight())
print("Second Triangle area:" ,t2.area())
 
t3 = t1 - t2 
print("Third Triangle Base:" , t3.getBase())
print("Third Triangle Height:" , t3.getHeight())
print("Third Triangle area:" ,t3.area())     
```

```python
#TASK no 8 
class Coordinates:

  def __init__(self,x, y):
    self.x = x
    self.y = y
  
  def __sub__(self, other):
    temp = self.x - other.x
    temp_2 = self.y-other.y
    return Coordinates(temp, temp_2)

  def __eq__(self, other):
    if self.x == other.x and self.y==other.y:
      return "The calculated coordinates are the same"
    else:
      return "The calculated coordinates are NOT the same."
      
  def __mul__(self, other):
    temp = self.x * other.x
    temp_2 = self.y * other.y
    return Coordinates(temp, temp_2)

  def detail(self):
    return f"({self.x}, {self.y})"


p1 = Coordinates(int(input()),int(input()))
p2 = Coordinates(int(input()),int(input()))
p4 = p1 - p2
print(p4.detail())
p5 = p1 * p2
print(p5.detail())
point_check = (p4 == p5)
print(point_check)
```
