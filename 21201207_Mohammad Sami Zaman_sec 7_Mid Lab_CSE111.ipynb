```python
# SET B
class PlayerDB:
    def __init__(self, *args):
        self.n = ''
        if len(args) == 2:
            self.p = args[0]
            self.t = args[1]
        elif len(args) == 1:
            self.t = 'Bangladesh'
            self.p = args[0]
            print('Someday we will win the world cup.')
        for i in self.p:
            if i != '@':
                self.n = self.n + i
            elif i == '@':
                break
        print(f'{self.n} added to {self.t} playerDB!')
    def addCareerStatistics(self, team = None, number = None, year = None):
        if year == None:
            return('Sorry! Year information must be provided!')
        else:
            self.team = team
            self.num = number
            self.year = year
            self.data = self.n + "@" + self.team  + '@' + str(self.num) + str(self.year)
            self.data_2 = self.p + '@' + self.t
            return('Player database updated')
    def showPlayerInfo(self):
        print('Player information: ', [self.data_2, self.data])
    def showDB(self, *args):
        dict_1 = {}
        dict_2 = {}
        self.d = {self.t : [self.data_2, self.data]}
        dict_2 = {args[0].t : [args[0].data_2, args[0].data]}
        for i in (args):
            dict_1= {(i.t) : [i.data_2, i.data]}
        print('All player database: ')
        print(self.d)
        print(dict_2)
        print(dict_1)

print("Player database activated. Have fun!")
print("1=========================")
Messi = PlayerDB("Messi@Ligue 1", "Argentina")
print("2=========================")
Jamal = PlayerDB("Jamal@BFPL")
print("3=========================")
print(Messi.addCareerStatistics("PSG", 40, 2021))
print("4=========================")
print(Jamal.addCareerStatistics("BFPL", 20))
print("5=========================")
print(Jamal.addCareerStatistics("BFPL", 20, 2020))
print("6=========================")
Messi.showPlayerInfo()
print("7=========================")
Jamal.showPlayerInfo()
print("8=========================")
Romero = PlayerDB("Romero@EPL", "Argentina")
print("9=========================")
print(Romero.addCareerStatistics("Tottenham", 10, 2021))
print("10========================")
Romero.showPlayerInfo()
print("11========================")
Romero.showDB(Messi, Jamal)
print("12========================")
```
