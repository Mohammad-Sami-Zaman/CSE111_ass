```python
#TASK no 1
class RealNumber:
    
    def __init__(self, r=0):
        self.__realValue = r  
    def getRealValue(self):
        return self.__realValue
    def setRealValue(self, r):
        self.__realValue = r
    def __str__(self):
        return 'RealPart: '+str(self.getRealValue())    

# WRITE HERE
class ComplexNumber(RealNumber):
    def __init__(self, real = 1, im = 1):
        RealNumber.__init__(self, float(real))
        self.__im = float(im)
    def __str__(self):
        return f'{RealNumber.__str__(self)} \n ImaginaryPart: {self.__im}'

cn1 = ComplexNumber()
print(cn1)
print('---------')
cn2 = ComplexNumber(5,7)
print(cn2)
```

```python
#TASK no 2
class RealNumber:
    def __init__(self, number=0):
        self.number = number
    def __add__(self, anotherRealNumber):
        return self.number + anotherRealNumber.number
    def __sub__(self, anotherRealNumber):
        return self.number - anotherRealNumber.number
    def __str__(self):
        return str(self.number)
#WRITE HERE
class ComplexNumber(RealNumber):
    def __init__(self, real , im):
      if type(real) == int:
        RealNumber.__init__(self, real)
      else:
        RealNumber.__init__(self, real.number)
      self. im = im
    def __add__(self,other):
        r = RealNumber.__add__(self, (other))
        im = self.im + other.im
        return ComplexNumber(r, im)
    def __sub__(self,other):
        r = RealNumber.__sub__(self, (other))
        im = self.im - other.im
        return ComplexNumber(r, im)
    def __str__(self):
      if self.im < 0:
        return f'{RealNumber.__str__(self)} {self.im}i'
      else:
        return f'{RealNumber.__str__(self)} + {self.im}i'
r1 = RealNumber(3)
r2 = RealNumber(5)
print(r1+r2)
cn1 = ComplexNumber(2, 1)
print(cn1)
cn2 = ComplexNumber(r1, 5)
print(cn2)
cn3 = cn1 + cn2
print(cn3)
cn4 = cn1 - cn2
print(cn4)
```

```python
#TASKNO 5
class Exam:
    def __init__(self,marks):
        self.marks = marks
        self.time = 60
        
    def examSyllabus(self):
        return "Maths , English"
    def examParts(self):
        return "Part 1 - Maths\nPart 2 - English\n"
#WRITE HERE
class ScienceExam(Exam):
    def __init__(self, *args):
      Exam.__init__(self, args[0])
      self.time = args[1]
      if len(args) == 4:
        var = Exam.examSyllabus(self).split(' , ')
        self.lst = var + [args[2], args[3]]
      else:
        var = Exam.examSyllabus(self).split(' , ')
        self.lst = var + [args[2], args[3], args[4]]
    def __str__(self):
      return(f'Marks: {self.marks} Time: {self.time} minutes Number of Parts: {len(self.lst)}')
    def examSyllabus(self):
      if len(self.lst) == 4:
        return f'{self.lst[0]}, {self.lst[1]}, {self.lst[2]}, {self.lst[3]}'
      else:
        return f'{self.lst[0]}, {self.lst[1]}, {self.lst[2]}, {self.lst[3]}, {self.lst[4]}'
    def examParts(self):
        print(Exam.examParts(self), end = '')
        for i in range(2, len(self.lst)):
            print(f'part {i+1} - {self.lst[i]}')
        return ''

    
engineering = ScienceExam(100,90,"Physics","HigherMaths")
print(engineering)
print('----------------------------------')
print(engineering.examSyllabus())
print(engineering.examParts())
print('==================================') 
architecture = ScienceExam(100,120,"Physics","HigherMaths","Drawing")
print(architecture)
print('----------------------------------')
print(architecture.examSyllabus())
print(architecture.examParts())
```

```python
#TASK no 6
class Shape3D:

  pi = 3.14159
  def __init__(self, name = 'Default', radius = 0):
    self._area = 0
    self._name = name
    self._height = 'No need'
    self._radius = radius   

  def calc_surface_area(self):
    return 2 * Shape3D.pi * self._radius

  def __str__(self):
      return "Radius: "+str(self._radius)
#WROTE HERE
class Sphere(Shape3D):
    def __init__(self, name, radius):
        Shape3D.__init__(self, name, radius)
        print(f'Shape name: {self._name}, Area Formula: 4 * pi * r * r')
    def calc_surface_area(self):
        self._area = Shape3D.calc_surface_area(self) * 2 * self._radius
    def __str__(self):
        return f'Radius: {self._radius}, Height: {self._height}\nArea: {self._area}'
class Cylinder(Shape3D):
    def __init__(self, name, radius, height):
        Shape3D.__init__(self, name, radius)
        self._height = height
        print(f'Shape name: {self._name}, Area Formula: 2 * pi * r * (r + h)')
    def calc_surface_area(self):
        self._area = Shape3D.calc_surface_area(self) * (self._radius + self._height)
    def __str__(self):
        return f'Radius: {self._radius}, Height: {self._height}\nArea: {self._area}' 

sph = Sphere('Sphere', 5)
print('----------------------------------')
sph.calc_surface_area()
print(sph)
print('==================================')
cyl = Cylinder('Cylinder', 5, 10)
print('----------------------------------')
cyl.calc_surface_area()
print(cyl)
```

```python
#TASK no  7
class PokemonBasic:

  def __init__(self, name = 'Default', hp = 0, weakness = 'None', type = 'Unknown'):
    self.name = name
    self.hit_point = hp
    self.weakness = weakness
    self.type = type

  def get_type(self):
    return 'Main type: ' + self.type

  def get_move(self):
    return 'Basic move: ' + 'Quick Attack'

  def __str__(self):
    return "Name: " + self.name + ", HP: " + str(self.hit_point) + ", Weakness: " + self.weakness
#WRITE HERE
class PokemonExtra(PokemonBasic):
    def __init__(self, name, hp, weakness, type_1, type_2=None, var=''):
        PokemonBasic.__init__(self, name, hp, weakness, type_1)
        self.t2 = type_2
        self.var = var
    def get_type(self):
        if self.t2 != None:
            return f'{PokemonBasic.get_type(self)}, Secondary Type: {self.t2}'
        else:
            return f'{PokemonBasic.get_type(self)}'  
    def get_move(self):
       text = ''
       for i in self.var:
          text = text + i + ','
       if self.var == '':
          return(f'{PokemonBasic.get_move(self)}')
       else:
          return (f'{PokemonBasic.get_move(self)}\nOther move: {text[0:len(text)-1]}')
        


print('\n------------Basic Info:--------------')
pk = PokemonBasic()
print(pk)
print(pk.get_type())
print(pk.get_move())

print('\n------------Pokemon 1 Info:-------------')
charmander = PokemonExtra('Charmander', 39, 'Water', 'Fire')
print(charmander)
print(charmander.get_type())
print(charmander.get_move())

print('\n------------Pokemon 2 Info:-------------')
charizard = PokemonExtra('Charizard', 78, 'Water', 'Fire', 'Flying', ('Fire Spin', 'Fire Blaze'))
print(charizard)
print(charizard.get_type())
print(charizard.get_move())
```

```python
#TASK NO 10
class Department:
    def __init__(self, s):
        self.semester = s
        self.name = "Default"
        self.id = -1
        
    def student_info(self):
        print("Name:", self.name)
        print("ID:", self.id) 
        
    def courses(self, c1, c2, c3):
        print("No courses Approved yet!")
#WRITE HERE
class CSE(Department):
    def __init__(self, name, id ,semester):
        Department.__init__(self, semester)
        self.name = name
        self.id = id
    def courses(self, *args):
        self.c1 = args[0]
        self.c2 = args[1]
        self.c3 = args[2]
        print(f'Courses Approved to this CSE Student in {self.semester} semester:')
        print(f'{self.c1}\n{self.c2}\n{self.c3}')

class EEE(Department):
    def __init__(self, name, id ,semester):
        Department.__init__(self, semester)
        self.name = name
        self.id = id
    def courses(self, *args):
        self.c1 = args[0]
        self.c2 = args[1]
        self.c3 = args[2]
        print(f'Courses Approved to this EEE Student in {self.semester} semester:')
        print(f'{self.c1}\n{self.c2}\n{self.c3}')
#==========================================================
s1 = CSE("Rahim", 16101328,"Spring2016")
s1.student_info()
s1.courses("CSE110", "MAT110", "ENG101")
print("==================")
s2 = EEE("Tanzim", 18101326, "Spring2018")
s2.student_info()
s2.courses("Mat110", "PHY111", "ENG101")
print("==================")
s3 = CSE("Rudana", 18101326, "Fall2017")
s3.student_info()
s3.courses("CSE111", "PHY101", "MAT120")
print("==================")
s4 = EEE("Zainab", 19201623, "Summer2019")
s4.student_info()
s4.courses("EEE201", "PHY112", "MAT120")
```
